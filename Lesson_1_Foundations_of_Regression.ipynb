## Part 1: NumPy Basics

NumPy allows us to work efficiently with numerical arrays.

Vectorised operations allow us to apply calculations to entire datasets at once.

```python
import numpy as np

hours = np.array([1, 2, 3, 4, 5])
print(hours)

print('Add 2 hours:', hours + 2)
print('Average study time:', np.mean(hours))
```

## Part 2: Visualising Data with Matplotlib

Before training a model, we visualise the relationship between variables.

Regression works best when the relationship appears linear.

```python
import matplotlib.pyplot as plt

scores = np.array([50, 55, 65, 70, 80])

plt.scatter(hours, scores)
plt.xlabel('Hours Studied')
plt.ylabel('Test Score')
plt.title('Study Time vs Test Score')
plt.show()
```

## Part 3: Linear Regression with Scikit-Learn

Linear regression attempts to model the relationship:
y = mx + b

This is supervised learning:
- X = input
- y = output


```python
from sklearn.linear_model import LinearRegression

X = hours.reshape(-1, 1)
y = scores

model = LinearRegression()
model.fit(X, y)

prediction = model.predict([[6]])
print('Predicted score for 6 hours:', prediction)
```

```python
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel('Hours Studied')
plt.ylabel('Test Score')
plt.title('Linear Regression Model')
plt.show()
```
